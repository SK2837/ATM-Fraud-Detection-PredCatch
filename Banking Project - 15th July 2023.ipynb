```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 70, 80, 90, 100],
    'criterion': ['gini', 'entropy'],
    'max_depth': [1, 2, 3, 4, 5],
    'min_samples_split': [2, 3, 5, 6, 7],
    'min_samples_leaf': [1, 2, 3, 4, 5],
    'min_weight_fraction_leaf': [0.0, 0.1, 0.5, 0.9, 1.0],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_leaf_nodes': [0, 1, 2, 3, 4],
    'min_impurity_decrease': [0.0, 0.1, 0.5, 0.9],
    'bootstrap': [True, False],
    'oob_score': [False, True],
    'max_samples': [1, 5, 10, 6, 8]
}

# Create the RandomizedSearchCV object
random_forest_hyper = RandomizedSearchCV(estimator=RandomForestClassifier(), param_distributions=param_grid, n_iter=10, cv=10)

```

# Model Building

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test= train_test_split(x_train_scaler, y_train_over , test_size=0.3, random_state=101)
```

```python
random_forest_hyper.fit(x_train, y_train)
```

```python
random_forest_hyper.best_params_
```

```python
{'oob_score': False,
 'n_estimators': 70,
 'min_weight_fraction_leaf': 0.1,
 'min_samples_split': 3,
 'min_samples_leaf': 3,
 'min_impurity_decrease': 0.1,
 'max_samples': 10,
 'max_leaf_nodes': 2,
 'max_features': 'auto',
 'max_depth': 4,
 'criterion': 'gini',
 'bootstrap': True}
```

```python
final_rf_model = RandomForestClassifier(
    oob_score=False,
    n_estimators=70,
    min_weight_fraction_leaf=0.1,
    min_samples_split=3,
    min_samples_leaf=3,
    min_impurity_decrease=0.1,
    max_samples=10,
    max_leaf_nodes=2,
    max_features='auto',
    max_depth=4,
    criterion='gini',
    bootstrap=True
)
```

```python
final_rf_model.fit(x_train, y_train)
```

```python
y_pred_train = final_rf_model.predict(x_train)
y_pred_test = final_rf_model.predict(x_test)
```

```python
#evaluation
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
print('Train Accouracy Score',accuracy_score(y_train, y_pred_train))
print('***************************** '*10)
print('Train Accouracy Score',accuracy_score(y_test, y_pred_test))
```
